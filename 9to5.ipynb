As we already know that 2017 has the highest earnings. In 2017 the top 2 highest earning industries are Total private, private service - providing.

```sql
select i.industry_name
        from LaborStatisticsDB.dbo.series s join LaborStatisticsDB.dbo.january_2017 j2017  
        on j2017.series_id=s.series_id 
        join LaborStatisticsDB.dbo.industry as i
        on i.industry_code=s.industry_code WHERE
        j2017.[value]>(
                select avg(a.[value])
from LaborStatisticsDB.dbo.series as s,LaborStatisticsDB.dbo.annual_2016 as a
where s.series_id=a.series_id and s.data_type_code=82
)and i.industry_name is not null group by i.industry_name order by sum(j2017.[value]) desc

```
